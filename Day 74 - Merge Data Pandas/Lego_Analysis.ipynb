### Aggregate Data with the Python .agg() Function

Let's work out the number of different themes shipped by year. This means we have to count the number of unique theme_ids per calendar year.

```python
# the .agg() method takes a dictionary as an argument
themes_by_year = sets.groupby("year").agg({"theme_id": pd.Series.nunique})
```

```python
# rename the column
themes_by_year.rename(columns={"theme_id": "nr_themes"}, inplace=True)
themes_by_year.head()
```

**Challenge**: Plot the number of themes released by year on a line chart. Only include the full calendar years (i.e., exclude 2020 and 2021).

```python
plt.plot(themes_by_year.index[:-2], themes_by_year.nr_themes[:-2])
```

### Line Charts with Two Seperate Axes
#### If we would like to add the two graphics it wouldn't be informative because they have different scales. For that we have to create two separated axes.

```python
ax1 = plt.gca() # get the current axis
ax2 = ax1.twinx() # create another axis that share the same x-axis
ax1.plot(sets_by_year.index[:-2], sets_by_year.set_num[:-2], color="g")
ax2.plot(themes_by_year.index[:-2], themes_by_year.nr_themes[:-2], color="b")
# now the graphic has one X and two Y

ax1.set_xlabel("Years")
ax1.set_ylabel("Number of Sets", color="g")
ax2.set_ylabel("Number of Themes", color="b")
```

**Challenge**: Use the <code>.groupby()</code> and <code>.agg()</code> function together to figure out the average number of parts per set. How many parts did the average LEGO set released in 1954 compared to say, 2017?

```python
parts_per_set = sets.groupby("year").agg({"num_parts": pd.Series.mean})
parts_per_set.head()
```

```python
parts_per_set.tail()
```

### Merging (i.e., Combining) DataFrames based on a Key
#### Let's use the .merge() method to combine two separate DataFrames into one. The merge method works on columns with the same name in both DataFrames.


```python
theme_count = pd.DataFrame({"id": theme_count.index, "set_count": theme_count.values})
theme_count.head()
```

```python
# merge table theme_count with themes file we're the common name is ID
merge_data = pd.merge(theme_count, themes, on="id")
merge_data[:5]
```

### Creating a Bar Char

```python
plt.figure(figsize=(14,8))
plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=14)
plt.ylabel('Number of Sets', fontsize=14)
plt.xlabel('Theme Name', fontsize=14)

plt.bar(merge_data.name[:10], merge_data.set_count[:10], color="pink")
```
